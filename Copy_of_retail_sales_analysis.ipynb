**Retail Sales Analysis**

**Project Description**

This project involves performing an Exploratory Data Analysis (EDA) on a retail transaction dataset to uncover insights into customer demographics, purchasing behavior, and product trends. The dataset includes information such as transaction details, customer profiles, product categories, and financial metrics.The analysis includes data cleaning, statistical exploration, and visualization using Python libraries such as pandas, matplotlib, and seaborn. Key outcomes include demographic breakdowns, product category analysis, and time-based sales trends.

**Objective**

To identify patterns and trends in sales and customer preferences.
To analyze correlations between variables like quantity, pricing, and total sales.
To derive actionable insights for optimizing business operations and marketing strategies.

   visualizations.

 1.Clear understanding of the customer demographics with respect to age and gender.

Exploratory Data Analysis





```python
# Distrubution Analysis
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=20, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Total Amount'], kde=True, bins=20, color='orange')
plt.title ('Total Amount Distrubution')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

```python

# 4. Customer Demographics
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.show()
```

```python
# Age segmentation
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '65+']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
plt.figure(figsize=(10, 6))
df['Age Group'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# product Analysis
plt.figure(figsize=(12,6))
df['Product Category'].value_counts().plot(kind='bar', color='green')
plt.title ('Product Category Popularity')
plt.xlabel('Product Category')
plt.ylabel('Number of Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Product Category', y= 'Price per Unit', data=df)
plt.title('price per unit by product category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Analysis
plt.figure(figsize=(10,8))
corr = df[['Quantity', 'Price per Unit', 'Total Amount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Time based analysis
# convert 'Date' to date time format
if df['Date'].dtype == 'object':
    df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Grouping by date
sales_trend = df.groupby('Date')['Total Amount'].sum()
plt.figure(figsize=(12,6))
sales_trend.plot()
plt.title('Sales trend over time')
plt.xlabel('Date')
plt.ylabel('Total sales')
plt.show()
```

```python
# scatter Plot for Quantity vs Total Amount
plt.figure(figsize=(10,6))
sns.scatterplot(x='Quantity', y='Total Amount', data=df)
plt.title('Quantity vs Total Amount')
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.show()
```
