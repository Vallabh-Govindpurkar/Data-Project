**Data Analysis of Play Store Apps Dataset**




## Objective

The goal of this analysis is to understand user ratings and app performance on the Google Play Store. We aim to uncover trends related to app category, pricing, and user reviews.

## Dataset Description

- **Source**: kaggle
- **Rows for Apps Dataset**: 9659
- **Rows for User_Review Dataset**: 64294
- **Columns for Apps Dataset**: 14
- **Columns for User_Review Dataset** : 5

- **Key Features for Apps Dataset**:

- `App`: The name of the application.
- `Category`: The category the app belongs to.
- `Rating`: User rating of the app.
- `Installs`: Number of times the app has been installed.

- **Key Features for User_Reviews Dataset**
- Translated_Reviews- Translated Reviews for each app
- Sentiment- sentiment for each app
- Sentiment_Polarity- Sentiment Polarity for each each app
- Sentiment_Subjectivity- Sentiment Subjectivity for each app

- **Import Libraries**
- Python libraries used
- Pandas: For data manipulation and cleaning.
- Matplotlib/Seaborn: For creating visualizations.

1. Top Category by Apps

```python
# Count the number of apps in each category
top_category = df1['Category'].value_counts().head(1)
print("Category with the most apps:")
print(top_category)

# Visualize the top categories
top_categories = df1['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 App Categories by Number of Apps')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.show()
```

2. Ratings by Type (Free vs Paid):

```python
# Group by 'Type' and calculate the average rating
ratings_by_type = df1.groupby('Type')['Rating'].mean()
print("Average Ratings by Type (Free vs Paid):")
print(ratings_by_type)

# Visualize Ratings by Type
sns.barplot(x=ratings_by_type.index, y=ratings_by_type.values, palette='coolwarm')
plt.title('Average Rating: Free vs Paid Apps')
plt.xlabel('Type')
plt.ylabel('Average Rating')
plt.ylim(0, 5)  # Ratings are typically between 0 and 5
plt.show()
```

```python
# Drop rows with missing reviews
df2.dropna(subset=['Translated_Review'],inplace = True)
```

```python
# Fill missing numerical values with appropriate values (e.g., mean for Sentiment_Polarity)
df2['Sentiment_Polarity'].fillna(df2['Sentiment_Polarity'].mean(), inplace=True)
df2['Sentiment_Subjectivity'].fillna(df2['Sentiment_Subjectivity'].mean(), inplace=True)
```

```python
df2.isnull().sum()
```

Sentiment Distrubution

```python
# Count the number of reviews for each sentiment
sentiment_counts = df2['Sentiment'].value_counts()

# Visualize the distribution of sentiments
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

Sentiment Polarity And subjectivity

```python
# Visualize sentiment polarity
sns.histplot(df2['Sentiment_Polarity'], bins=20, kde=True, color='blue')
plt.title('Distribution of Sentiment Polarity')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()

# Visualize sentiment subjectivity
sns.histplot(df2['Sentiment_Subjectivity'], bins=20, kde=True, color='orange')
plt.title('Distribution of Sentiment Subjectivity')
plt.xlabel('Sentiment Subjectivity')
plt.ylabel('Frequency')
plt.show()
```

Sentiment analysis by app

```python
# Calculate average sentiment polarity and subjectivity for each app
app_sentiment = df2.groupby('App')[['Sentiment_Polarity', 'Sentiment_Subjectivity']].mean()

# Top 10 apps with the highest polarity (most positive sentiment)
top_positive_apps = app_sentiment.sort_values(by='Sentiment_Polarity', ascending=False).head(10)
print("Top 10 Apps with the Most Positive Sentiment:")
print(top_positive_apps)

# Visualize the results
top_positive_apps['Sentiment_Polarity'].plot(kind='bar', color='green')
plt.title('Top 10 Apps with Positive Sentiment')
plt.xlabel('App')
plt.ylabel('Average Sentiment Polarity')
plt.show()
```

Most Subjective or Objective Reviews

```python
# Find the most subjective and objective reviews
most_subjective_reviews = df2.nlargest(5, 'Sentiment_Subjectivity')[['Translated_Review', 'Sentiment_Subjectivity']]
most_objective_reviews = df2.nsmallest(5, 'Sentiment_Subjectivity')[['Translated_Review', 'Sentiment_Subjectivity']]

print("Most Subjective Reviews:")
print(most_subjective_reviews)

print("\nMost Objective Reviews:")
print(most_objective_reviews)
```
